```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos del ejercicio 1
x1 = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y1 = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# a. Polinomio de grado 1 (lineal)
coeff1 = np.polyfit(x1, y1, 1)
p1 = np.poly1d(coeff1)
error1 = np.sum((y1 - p1(x1))**2)

# b. Polinomio de grado 2 (cuadrático)
coeff2 = np.polyfit(x1, y1, 2)
p2 = np.poly1d(coeff2)
error2 = np.sum((y1 - p2(x1))**2)

# c. Polinomio de grado 3 (cúbico)
coeff3 = np.polyfit(x1, y1, 3)
p3 = np.poly1d(coeff3)
error3 = np.sum((y1 - p3(x1))**2)

# d. Función exponencial be^ax
def exp_func(x, a, b):
    return b * np.exp(a * x)
popt_exp, pcov_exp = curve_fit(exp_func, x1, y1)
error_exp = np.sum((y1 - exp_func(x1, *popt_exp))**2)

# e. Función potencia bx^a
def power_func(x, a, b):
    return b * x**a
popt_pow, pcov_pow = curve_fit(power_func, x1, y1)
error_pow = np.sum((y1 - power_func(x1, *popt_pow))**2)

# Graficar todos los ajustes
plt.figure(figsize=(12, 8))
plt.scatter(x1, y1, color='red', label='Datos originales')

x_vals = np.linspace(min(x1), max(x1), 100)
plt.plot(x_vals, p1(x_vals), label=f'Lineal (Error: {error1:.2f})')
plt.plot(x_vals, p2(x_vals), label=f'Cuadrático (Error: {error2:.2f})')
plt.plot(x_vals, p3(x_vals), label=f'Cúbico (Error: {error3:.2f})')
plt.plot(x_vals, exp_func(x_vals, *popt_exp), 
         label=f'Exponencial (Error: {error_exp:.2f})')
plt.plot(x_vals, power_func(x_vals, *popt_pow), 
         label=f'Potencia (Error: {error_pow:.2f})')

plt.title('Ajustes por mínimos cuadrados - Ejercicio 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Mostrar ecuaciones y errores
print("a. Lineal: y =", p1, f"\nError: {error1:.2f}")
print("b. Cuadrático: y =", p2, f"\nError: {error2:.2f}")
print("c. Cúbico: y =", p3, f"\nError: {error3:.2f}")
print(f"d. Exponencial: y = {popt_exp[1]:.2f}e^({popt_exp[0]:.2f}x) \nError: {error_exp:.2f}")
print(f"e. Potencia: y = {popt_pow[1]:.2f}x^{popt_pow[0]:.2f} \nError: {error_pow:.2f}")
```

```python
# Datos del ejercicio 2
x2 = np.array([0.2, 0.3, 0.6, 0.9, 1.1, 1.3, 1.4, 1.6])
y2 = np.array([0.050446, 0.098426, 0.33277, 0.72660, 1.0972, 1.5697, 1.8487, 2.5015])

# Ajustes para el ejercicio 2
# a. Lineal
coeff1_2 = np.polyfit(x2, y2, 1)
p1_2 = np.poly1d(coeff1_2)
error1_2 = np.sum((y2 - p1_2(x2))**2)

# b. Cuadrático
coeff2_2 = np.polyfit(x2, y2, 2)
p2_2 = np.poly1d(coeff2_2)
error2_2 = np.sum((y2 - p2_2(x2))**2)

# c. Cúbico
coeff3_2 = np.polyfit(x2, y2, 3)
p3_2 = np.poly1d(coeff3_2)
error3_2 = np.sum((y2 - p3_2(x2))**2)

# d. Exponencial
popt_exp2, pcov_exp2 = curve_fit(exp_func, x2, y2)
error_exp2 = np.sum((y2 - exp_func(x2, *popt_exp2))**2)

# e. Potencia
popt_pow2, pcov_pow2 = curve_fit(power_func, x2, y2)
error_pow2 = np.sum((y2 - power_func(x2, *popt_pow2))**2)

# Graficar
plt.figure(figsize=(12, 8))
plt.scatter(x2, y2, color='red', label='Datos originales')

x_vals2 = np.linspace(min(x2), max(x2), 100)
plt.plot(x_vals2, p1_2(x_vals2), label=f'Lineal (Error: {error1_2:.4f})')
plt.plot(x_vals2, p2_2(x_vals2), label=f'Cuadrático (Error: {error2_2:.4f})')
plt.plot(x_vals2, p3_2(x_vals2), label=f'Cúbico (Error: {error3_2:.4f})')
plt.plot(x_vals2, exp_func(x_vals2, *popt_exp2), 
         label=f'Exponencial (Error: {error_exp2:.4f})')
plt.plot(x_vals2, power_func(x_vals2, *popt_pow2), 
         label=f'Potencia (Error: {error_pow2:.4f})')

plt.title('Ajustes por mínimos cuadrados - Ejercicio 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Mostrar ecuaciones y errores
print("a. Lineal: y =", p1_2, f"\nError: {error1_2:.6f}")
print("b. Cuadrático: y =", p2_2, f"\nError: {error2_2:.6f}")
print("c. Cúbico: y =", p3_2, f"\nError: {error3_2:.6f}")
print(f"d. Exponencial: y = {popt_exp2[1]:.4f}e^({popt_exp2[0]:.4f}x) \nError: {error_exp2:.6f}")
print(f"e. Potencia: y = {popt_pow2[1]:.4f}x^{popt_pow2[0]:.4f} \nError: {error_pow2:.6f}")
```
